```python
from bs4 import BeautifulSoup
import requests
import re
import time
import os
```

```python
category_code = {
    "economy":"22",
    "politics":"21",
    "society": "24",
    "art": "25",
    "sports":"26",
    "world":"27",
    "foreign":"28",
    "technology":"33",
    "health":"31",
    "opinion":"81",
    "interview":"82"    
}
def extract_and_save(short_url,category):
    time.sleep(2)
    news_url = "https://www.nagariknews.com" + short_url
    print(news_url)
    content = requests.get(news_url).text
    tree = BeautifulSoup(content,"lxml")
    news_text = tree.find_all("div",{"class":"tag news-content"})[0]
    with open(category +"/" + short_url.strip("/news/")+".txt","w") as file:
        for i in news_text.find_all("p"):
            file.write(i.text)
            file.write("\n")
```

```python
def nagarikParse(category,start_page = 1,limit = None):
    if not os.path.isdir(category):
        os.mkdir(category)
    if limit is not None:
        count =  0
    parsed_till_now = set()
    page = start_page
    base_url = "https://www. .com" + "/category/" + category_code[category]
    url = base_url 
    while url is not None:
        time.sleep(2)
        print(url)
        content = requests.get(url).content
        tree = BeautifulSoup(content,"lxml")
        
        #Extracting Relevent Links

        # Seperating Out the column div that Contains news
        column_tree = tree.find_all('div',{"class":"col-sm-9 left-on"})[0]
        column_links = column_tree.find_all('a')
        column_hrefs = [j['href'] 
                            for j in column_links 
                                if j['href'] not in parsed_till_now]
        hrefs = set(column_hrefs)
        
        #A seperate row contains new news present in the first page but not in the rest of pages
        row_tree = tree.find_all('div',{"class":"col-sm-3 part-ent"})
        row_hrefs = [j['href']
                 for i in row_tree
                    for j in i.find_all('a')
                         if j['href'] not in parsed_till_now
        ]
        hrefs.update(row_hrefs)
        links_first_row = []
        
        #This is the set of next links
        news_links = filter(lambda x:(not re.match("/news",x) is None) and re.match("/news/author",x) is None  ,hrefs)
        
        ##Extract Text and Save to File
        for i in news_links:
            extract_and_save(i,category)
            if limit is not None:
                count += 1
                if count > limit:
                    print("Limit Reached")
                    return
        parsed_till_now.update(news_links)
        ##Pagination
        next_page = column_tree.find_all("ul",{"class":"pagination"})
        if len(next_page[0].find_all('a',{"rel":"next"})) != 0:
            page+=1
            url = base_url + "?page=" + str(page)
        else:
            url = None
            print("Parse Complete")
```

```python
for i in category_code:
    if i in (  "politics","foreign","economy"):
        continue
    nagarikParse(i,start_page = 1 ,limit=510)
    
```
